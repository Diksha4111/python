```python
"""append() and extend() takes only one argument.
The difference between append() and extend() is that append() will add the whole list at the end of the list 
but extend() adds the elements passed in the as an argument at the end of the list."""

x=[10, 20, 30, 40, 50]
x.append(['hi','bye'])
print(x)
x.extend(['hi','bye'])
print(x)
```
