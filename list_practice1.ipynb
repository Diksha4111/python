```python
""" join() function joins the list of strings with a separator. 
join() can take string,list,tuple,dictionary and set as an argument."""

y=['a','b','c','d']
k='--'
k=k.join(y)
k
```

```python
# append(element) function inserts an element passed as an argument at the end of the list. It takes element as an argument.

x=[9,5,7,2,1,0,5,3,11]
x.append(100)
x
```

```python
"""pop(index_value) takes index value as an argument and 
removes the element from the end of the list i.e. the last element of the list if nothing is given as an argument."""

x=[9, 5, 7, 2, 1, 0, 5, 3, 11, 100]
x.sort()
x.pop()
x
```

```python
''' count(element) counts how many times the element which is passed as an argument occurs in the list. 
It takes element as an argument. '''

a=[1,6,8,0,4,2,6,1,1,0,9]
a.count(1)
```

```python
#list slicing---list[start:stop:step]. Start and step are optional. By default start=0 and step=1.

c=[1,8,4,0,'a','$','d',100]
print(c[5])
print(c[::])
print(c[1:3])
print(c[0:7:2])
print(c[0:4:-1])
print(c[2:6])
print(c[-1:-9:-1])
print(c[-1:3])
print(c[-1:3:-1])
print(c[-1:2:-2])
```

```python
#len() is used to find the length of whichever is passed as an argument i.e whether a string, list or tuple.

a=['d','i','k','s','h','a',1,1,1]
l=len(a)
for i in range(l):
    print(a[i],end="-")
```
