```python
#Exp2_1

n=int(input("How many student?"))
s=int(input("How many subjects?"))
m=[]
details={}
for i in range(n):
    name=input("Enter name:")
    for i in range(s):
        marks=int(input("Enter marks:"))
        m.append(marks)
    details[name]=m

sum=0
ok=input("Enter strudent name whom avg marks you want to retrive:")
for i in details.items():
    if(i[0]==ok):
        for j in range(s):
            sum+=i[1][j]


avg=sum/s
print("The avg marks of",ok," are:",'%.2f'%avg)#"Avg marks of",i[0],"student are:",avg)
    #sum=0
#print("avg=",avg)
```

```python
#Exp2_2

x=[]
n=int(input("How many students?"))
for i in range(n):
    marks=int(input("Enter marks:"))
    x.append(marks)

x.sort()
for i in x:
    x.remove(min(x))
    sec_max=i
    if(sec_max==min(x)):
        break
sec_max=min(x)
print("The second max is:",sec_max)    
```
